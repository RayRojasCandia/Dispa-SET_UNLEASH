    4. Missed Values Fulfilling
</div>
<div style="text-align: justify; margin-left: 0.0em; font-weight: unbold; font-size: 14px; font-family: TimesNewRoman; color:skyblue">
Because of the data from the original sources, there are some time points that might have the value of '0'.
<br>
This missed points are completed using the data from previous years.
</div>

```python
def adjust_year_in_timestamp(timestamp, new_year):
    """
    Adjust the year in the timestamp to the new year, keeping the same date and time structure.
    """
    dt = datetime.strptime(timestamp, '%Y-%m-%d %H:%M:%S+00:00')
    return dt.replace(year=new_year).strftime('%Y-%m-%d %H:%M:%S+00:00')

def find_value_in_other_files(target_year, target_timestamp, dataframes, data_years_list, search_previous=True):
    """
    Search for the target timestamp in previous or next CSV files to find a value greater than 0.
    If search_previous is True, look in previous years, otherwise look in future years.
    """
    current_year_index = data_years_list.index(target_year)
    
    if search_previous:
        search_range = range(current_year_index - 1, -1, -1)  # Previous years
    else:
        search_range = range(current_year_index + 1, len(data_years_list))  # Future years
    
    for index in search_range:
        other_year = data_years_list[index]
        df_other = dataframes[other_year]
        
        # Adjust the timestamp to match the year of the other file
        adjusted_timestamp = adjust_year_in_timestamp(target_timestamp, other_year)
        
        # Look for the adjusted timestamp in the first column (index)
        if adjusted_timestamp in df_other.index:
            value = df_other.loc[adjusted_timestamp, df_other.columns[1]]  # Second column value
            if value > 0:  # If value is greater than 0, return it
                print(f"Found value {value} in {other_year}.csv for timestamp {adjusted_timestamp}")
                return value
        else:
            print(f"{other_year}.csv does not contain the timestamp {adjusted_timestamp}, skipping...")
    
    return None  # If no value is found, return None

def process_csv_files(data_years_list, load_realtime_folder_path):
    """
    Process all CSV files in the specified folder structure.
    """
    dataframes = {year: None for year in data_years_list}
    
    # Load all relevant CSV files into memory
    for year in data_years_list:
        file_path = os.path.join(load_realtime_folder_path, f"{year}.csv")
        if os.path.exists(file_path):
            df = pd.read_csv(file_path)
            df.columns = df.columns.str.strip()  # Clean column names
            df.set_index(df.columns[0], inplace=True)  # Set timestamp as index
            dataframes[year] = df
            print(f"Loaded file: {file_path}")
        else:
            print(f"File {file_path} does not exist, skipping...")

    # Now process the loaded DataFrames
    for root, dirs, files in os.walk(load_realtime_folder_path):
        for file in files:
            if file.endswith('.csv'):
                year = int(file.split('.')[0])
                if year in data_years_list:
                    df = dataframes[year]
                    
                    # Loop through rows where the second column has a value of '0'
                    for idx in df[df[df.columns[1]] == 0].index:
                        target_timestamp = idx  # Use index directly as timestamp
                        print(f"Found '0' value at row {idx}, timestamp: {target_timestamp}. Searching for replacement...")
                        
                        new_value = None
                        # Try searching in previous years first
                        new_value = find_value_in_other_files(year, target_timestamp, dataframes, data_years_list, search_previous=True)
                        
                        if new_value is None:  # If no value was found in previous years, search future years
                            new_value = find_value_in_other_files(year, target_timestamp, dataframes, data_years_list, search_previous=False)
                        
                        if new_value is not None:
                            df.at[idx, df.columns[1]] = new_value
                            print(f"Replaced '0' value at row {idx} with {new_value}.")
                        else:
                            print(f"No replacement found for row {idx}, keeping original '0' value.")
                    
                    # Save the modified DataFrame back to the CSV file
                    df.to_csv(os.path.join(root, file), index=True)
                    print(f"File {file_path} updated successfully.")
                else:
                    print(f"Year {year} not in data_years_list, skipping {file_path}.")

process_csv_files(data_years_list, load_realtime_folder_path)
```

<div style="background-color: black;">
<div style="text-align: justify; margin-left: 0.0em; font-weight: unbold; font-size: 14px; font-family: TimesNewRoman; color:skyblue">
There remain some missed points that do not have any value even for all the analized years.
<br>
This missed points are completed using the data from the closest time step in the same hour inside the analized year.
</div>

```python
def is_leap_year(year):
    """Check if a year is a leap year."""
    return year % 4 == 0 and (year % 100 != 0 or year % 400 == 0)

def adjust_year_in_timestamp(timestamp, new_year):
    """
    Adjust the year in the timestamp to the new year, keeping the same date and time structure.
    If the date is February 29 and the new year is not a leap year, adjust to February 28.
    """
    dt = datetime.strptime(timestamp, '%Y-%m-%d %H:%M:%S+00:00')
    
    # If the original date is February 29 and the new year is not a leap year, adjust to February 28
    if dt.month == 2 and dt.day == 29 and not is_leap_year(new_year):
        return dt.replace(year=new_year, day=28).strftime('%Y-%m-%d %H:%M:%S+00:00')
    
    return dt.replace(year=new_year).strftime('%Y-%m-%d %H:%M:%S+00:00')

def find_next_valid_value(df, start_idx, target_time):
    """
    Look for the next non-zero value in the same file that has the same hour as the target_time.
    """
    target_hour = datetime.strptime(target_time, '%Y-%m-%d %H:%M:%S+00:00').hour
    
    for idx in range(start_idx + 1, len(df)):
        next_timestamp = df.iloc[idx, 0]
        next_value = df.iloc[idx, 1]
        
        # Extract the hour from the next timestamp
        next_hour = datetime.strptime(next_timestamp, '%Y-%m-%d %H:%M:%S+00:00').hour
        
        # If the hour matches and the value is greater than 0, return the value
        if next_hour == target_hour and next_value > 0:
            return next_value
    
    return None  # No valid value found in the current file

def process_csv_file(file_path, year):
    """
    Process a single CSV file, looking for 0 values and replacing them with the next valid value.
    """
    if os.path.exists(file_path):
        print(f"Processing file: {file_path}")
        
        # Load the CSV without headers
        df = pd.read_csv(file_path, header=None)
        
        # Assign default column names (since there are no headers)
        df.columns = ['Timestamp', 'Value']
        
        # Loop through rows where the second column (Value) has a value of '0'
        for idx, row in df[df['Value'] == 0].iterrows():
            target_timestamp = row['Timestamp']  # First column (Timestamp)
            
            # Find the next valid value in the current file
            new_value = find_next_valid_value(df, idx, target_timestamp)
            
            if new_value is not None:
                # Update the value in the DataFrame
                df.at[idx, 'Value'] = new_value
                print(f"Replaced '0' value at row {idx} with {new_value}.")
            else:
                print(f"No valid replacement found for row {idx}, keeping original '0' value.")
        
        # Save the modified DataFrame back to the CSV file (without headers)
        df.to_csv(file_path, index=False, header=False)
        print(f"File {file_path} updated successfully.")
    else:
        print(f"File {file_path} does not exist, skipping...")

def process_csv_files_in_directory(directory_path, data_years_list):
    """
    Process CSV files in a specific sub-directory (like '1h', '30min', or '15min').
    """
    for year in data_years_list:
        file_path = os.path.join(directory_path, f"{year}.csv")
        process_csv_file(file_path, year)

def process_all_csv_files(load_realtime_folder_path, data_years_list):
    """
    Recursively go through all sub-folders in the load_realtime_folder_path, find '1h', '30min', and '15min' folders,
    and process the CSV files inside.
    """
    for root, dirs, files in os.walk(load_realtime_folder_path):
        for subdir in ['1h', '30min', '15min']:
            subdir_path = os.path.join(root, subdir)
            if os.path.exists(subdir_path):
                print(f"Processing CSV files in folder: {subdir_path}")
                process_csv_files_in_directory(subdir_path, data_years_list)
            else:
                print(f"Sub-folder {subdir} does not exist in {root}, skipping...")

# Example usage
load_realtime_folder_path = load_realtime_folder_path  # Replace with your actual directory path

process_all_csv_files(load_realtime_folder_path, data_years_list)
```
